# Phase 6c — Push Fine-Tuned TrOCR to Hugging Face Hub

One-time operation. Pushes the fine-tuned model saved in Drive to a
private Hugging Face Hub model repo so the Gradio app can load it.

Run this once after Phase 6a completes.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import subprocess, sys
def pip(pkg): subprocess.check_call([sys.executable, '-m', 'pip', 'install', pkg, '-q'])
pip('huggingface_hub')
print('Done')
```

```python
from pathlib import Path
from huggingface_hub import HfApi, login

DRIVE_ROOT   = Path('/content/drive/MyDrive/vision-ocr-accessibility-assistant')
FINETUNE_DIR = DRIVE_ROOT / 'models/trocr-finetuned/trocr-finetuned'

assert FINETUNE_DIR.exists(), f'Fine-tuned model not found at {FINETUNE_DIR}. Run 06a first.'
print(f'Model found at: {FINETUNE_DIR}')
print(f'Files: {list(FINETUNE_DIR.iterdir())}')
```

```python
# Login to Hugging Face
# Get your token from: https://huggingface.co/settings/tokens
# Use a token with write access
login()
```

```python
api     = HfApi()
REPO_ID = 'MoCamaraData/trocr-ocr-accessibility'

# Create repo if it does not exist
api.create_repo(
    repo_id=REPO_ID,
    repo_type='model',
    private = False,
    exist_ok=True
)
print(f'Repo ready: {REPO_ID}')

# Push fine-tuned weights
print('Uploading model weights...')
api.upload_folder(
    folder_path=str(FINETUNE_DIR),
    repo_id=REPO_ID,
    repo_type='model'
)
print(f'Model pushed to: https://huggingface.co/{REPO_ID} ✓')
```

```python
from pathlib import Path
from huggingface_hub import HfApi

DRIVE_ROOT   = Path('/content/drive/MyDrive/vision-ocr-accessibility-assistant')
FINETUNE_DIR = DRIVE_ROOT / 'models/trocr-finetuned/trocr-finetuned'
REPO_ID      = "MoCamaraData/trocr-ocr-accessibility"

assert FINETUNE_DIR.exists(), f"Model folder not found: {FINETUNE_DIR}"
print("Uploading from:", FINETUNE_DIR)
print("Files:", [p.name for p in FINETUNE_DIR.iterdir()])

api = HfApi()
api.upload_folder(
    folder_path=str(FINETUNE_DIR),
    repo_id=REPO_ID,
    repo_type="model",
    commit_message="Upload improved final fine-tuned TrOCR model",
)

print("Updated model uploaded successfully.")
```
